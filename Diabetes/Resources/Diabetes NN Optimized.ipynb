```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 9
hidden_nodes_layer2 = 1
hidden_nodes_layer3 = 3
hidden_nodes_layer4 = 9

beet_nn = tf.keras.models.Sequential()

# First hidden layer
beet_nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
beet_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
beet_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# fourth hidden layer
beet_nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))

# Output layer
beet_nn.add(tf.keras.layers.Dense(units=1, activation="relu"))

# Check the structure of the model
beet_nn.summary()
```

```python
#Compiling
beet_nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#training model
fit_model = beet_nn.fit(X_train_scaled,y_train,epochs=20)
```

```python
model_loss, model_accuracy = beet_nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
